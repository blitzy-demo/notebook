# User Presence and Cursor Awareness

This notebook demonstrates the user presence and awareness system in Jupyter Notebook v7's collaborative editing environment. When multiple users are working on the same notebook simultaneously, the presence system provides visual indicators showing who is active, where they are working, and what they are doing in real-time.

## Understanding User Presence

User presence awareness is implemented through the **Yjs awareness protocol**, which synchronizes user state information across all connected clients. This system tracks and displays:

* **Active collaborators** - List of users currently connected to the notebook
* **Cursor positions** - Real-time cursor locations within cells
* **Text selections** - Highlighted text selections made by each user
* **Active cells** - Which cell each user is currently focused on
* **User metadata** - Names, avatars, and assigned colors for identification

### Visual Indicators in the Interface

When collaboration is active, you'll see several visual elements in the notebook interface:

1. **Collaboration Toolbar**: Located at the top of the notebook, showing avatars of active users
2. **Cursor Indicators**: Colored cursors showing where other users are typing
3. **Selection Highlights**: Colored overlays showing text selected by other users
4. **Cell Focus Indicators**: Subtle border colors indicating which cells other users are working on
5. **User Activity Status**: Real-time indicators showing user connection and activity state

```python
# This is a sample code cell where cursor awareness is demonstrated
# When another user is editing this cell, you would see:
# - Their cursor position as a colored vertical line
# - Any text they select highlighted in their assigned color
# - The cell border may change color to indicate their focus

import numpy as np
import matplotlib.pyplot as plt

# Create sample data for demonstration
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Collaborative Data Analysis Example')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

## Yjs Awareness Protocol Integration

The presence system is built on top of the **Yjs awareness protocol**, which provides conflict-free, real-time synchronization of user state information. Here's how it works technically:

### Awareness State Structure

```python
# Conceptual representation of awareness state tracked by Yjs
# This is automatically managed by the collaboration system

awareness_state_example = {
    "user": {
        "name": "Alice Johnson",
        "id": "user_12345",
        "color": "#3B82F6",  # Blue color assigned to this user
        "avatar": "https://example.com/avatar/alice.jpg"
    },
    "cursor": {
        "cellId": "cell_67890",
        "position": 127,  # Character position within the cell
        "line": 5,        # Line number within the cell
        "column": 23      # Column position on the line
    },
    "selection": {
        "cellId": "cell_67890",
        "start": 120,     # Start position of selection
        "end": 135,       # End position of selection
        "text": "plt.show()"  # Selected text
    },
    "timestamp": 1672531200000,  # Last activity timestamp
    "activeCell": "cell_67890",  # Currently focused cell
    "connectionState": "connected"  # Connection status
}

print("Awareness state is synchronized across all connected clients")
print(f"User {awareness_state_example['user']['name']} is active in cell {awareness_state_example['activeCell']}")
```

### Real-time Synchronization Process

The awareness protocol operates through the following process:

1. **State Updates**: When a user moves their cursor, selects text, or changes focus, their client updates its local awareness state
2. **Broadcasting**: The updated state is immediately broadcast to all other connected clients via WebSocket
3. **Conflict Resolution**: Multiple simultaneous updates are merged using Yjs CRDT algorithms
4. **UI Rendering**: Each client receives the merged state and updates the visual indicators accordingly
5. **Cleanup**: Inactive users are automatically removed from the awareness state after a timeout period

```python
# Example of how awareness updates flow through the system
# This demonstrates the technical flow, not actual implementation code

def simulate_awareness_flow():
    """
    Simulates how awareness information flows in the collaborative system
    """
    
    # Step 1: User interaction triggers awareness update
    user_action = "cursor_move"
    new_position = 145
    
    print(f"Step 1: User performs {user_action} to position {new_position}")
    
    # Step 2: Local awareness state is updated
    local_awareness = {
        "cursor": {"position": new_position, "cellId": "current_cell"},
        "timestamp": 1672531250000
    }
    print(f"Step 2: Local awareness updated: {local_awareness}")
    
    # Step 3: Update is broadcast to other clients
    print("Step 3: Broadcasting update via Yjs WebSocket connection")
    
    # Step 4: Remote clients receive and apply the update
    print("Step 4: Other clients receive update and render cursor at new position")
    
    # Step 5: Visual indicators are updated in real-time
    print("Step 5: UI elements updated - cursor indicators, user lists, etc.")
    
    return "Awareness synchronization complete"

result = simulate_awareness_flow()
print(f"\nResult: {result}")
```

## Collaborative Editing Scenarios

The following scenarios demonstrate how presence awareness enhances collaborative workflows:

### Scenario 1: Multi-user Data Analysis

**Situation**: Three data scientists (Alice, Bob, and Carol) are collaboratively analyzing a dataset.

**Presence Indicators**:
- Alice's avatar appears in **blue** in the collaboration toolbar
- Bob's avatar appears in **green** with a cursor in the code cell below
- Carol's avatar appears in **purple** with text selection in a markdown cell

**What you would see**:
- Bob's green cursor moving through the pandas code as he types
- Carol's purple highlight over text she's reviewing in documentation
- Real-time updates as each person makes changes

```python
# In this cell, you might see Bob's green cursor here ↓
import pandas as pd

# Load the collaborative dataset
# Bob might be typing additional analysis code here
data = pd.read_csv('collaborative_dataset.csv')

# Alice might be working on data cleaning below
cleaned_data = data.dropna()

# Carol might be adding documentation comments
# Each user's contributions are visible in real-time
print(f"Dataset shape: {cleaned_data.shape}")
print(f"Collaborators can see each other's cursor positions and selections")
```

### Scenario 2: Code Review and Discussion

**Situation**: A senior developer is reviewing junior developer's code while providing guidance.

**Presence Benefits**:
- Reviewer can see exactly where the developer is focused
- Text selections help highlight areas under discussion
- Real-time cursor movement shows active engagement
- Multiple people can point to specific code sections simultaneously

```python
# Code review scenario - multiple cursors might be visible here
def complex_analysis_function(data, parameters):
    """
    Reviewer might select this docstring to discuss documentation
    Junior dev might have cursor in the function body
    """
    
    # Reviewer might highlight this line for discussion
    result = data.groupby('category').agg({
        'value': ['mean', 'std', 'count'],  # Junior dev might be editing here
        'date': 'max'
    })
    
    # Both users can see cursor positions during code review
    processed_result = result.reset_index()
    
    return processed_result

# Visual indicators help coordinate the review process
print("Code review enhanced by real-time presence awareness")
```

### Scenario 3: Teaching and Learning

**Situation**: An instructor is teaching a programming class with students following along.

**Educational Benefits**:
- Students can see instructor's cursor during live coding demonstrations
- Instructor can monitor student progress by observing their cursor positions
- Text selections help highlight important concepts
- Real-time feedback on student engagement and understanding

```python
# Educational scenario - instructor demonstrating concepts
# Students would see instructor's cursor moving through this code

def fibonacci_sequence(n):
    """
    Instructor might select this line to explain recursion
    Students' cursors might follow along in their own attempts
    """
    
    if n <= 1:
        return n  # Instructor highlights base case
    else:
        # Students might have cursors here as they type
        return fibonacci_sequence(n-1) + fibonacci_sequence(n-2)

# Demonstration with visual feedback
for i in range(10):
    fib_num = fibonacci_sequence(i)
    print(f"F({i}) = {fib_num}")
    
print("\nPresence awareness enhances the learning experience")
```
